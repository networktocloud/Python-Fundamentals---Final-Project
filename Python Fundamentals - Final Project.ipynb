```python
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
import seaborn as sns
```

```python
df_mpg = sns.load_dataset("mpg")
df_mpg
```

```python
#Hit button to sum the amount of cylinders and display
cylinders = df_mpg.cylinders
def add_me(obj):
    total = 0
    with output:
        for cyl in cylinders:
            total += cyl
        print("Total cylinders of {} button is: ".format(obj.description) + str(total))

output = widgets.Output()

show_total_cylinders = widgets.Button(description = "Add me", icon = "plus")

display(show_total_cylinders, output)
show_total_cylinders.on_click(add_me)
```
